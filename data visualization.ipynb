```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df=sns.load_dataset('iris')
iris_df
```

```python
iris_df['species'].value_counts()
```

```python
sns.scatterplot(x=iris_df['sepal_length'],y=iris_df['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris_df['sepal_length'],y=iris_df['petal_length'],hue=iris_df['species'])
plt.show()
```

```python
sns.boxplot(x=iris_df['species'],y=iris_df['petal_length'])
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'])
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='reg')
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='scatter',color='red')
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='kde')
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'],kind='hex')
plt.show()
```

```python
sns.countplot(iris_df['species'],color='green')
```

```python
sns.countplot(y=iris_df['species'],palette=sns.cubehelix_palette(),color='red')
```

```python
sns.distplot(iris_df.petal_length,kde=True)
```

```python
sns.barplot(data=iris_df,x="sepal_length",y="sepal_width",hue="species")
```

```python
corr=iris_df.corr()
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

```python
sns.pairplot(data=iris_df,hue="species",kind="scatter",palette="husl")
```
